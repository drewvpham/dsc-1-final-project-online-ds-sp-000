### 2c. Fill in the missing values

```python
# Since most waterfront is zero, fill in the waterfront missing values with 0, since houses are either with waterfront (1)or without (0).
df.loc[df.waterfront.isna(), 'waterfront'] = 0
df.waterfront = df.waterfront.astype('int')

# We have 5 unique values for view and we would like to use the median, for filing in the missing values. 
#In this case, the median is 0, which is also the mode for this column
df['view'].fillna(df['view'].median(), inplace=True )
df.view = df.view.astype('int')

# We will replace the NaN with 'missing' and the 0 value in the yr_renovated features
df.loc[df.yr_renovated.isna(), 'yr_renovated'] = 0

# Let's confirm that the missing values were filled in (verify increase in 0.0 totals)
display(df['view'].value_counts(), df['waterfront'].value_counts(), df['yr_renovated'].value_counts().nlargest(5))
```

### 4a. Preprocess the Dummy Variables

We need to transform the categorical variables before using them in regression models. 
We will be using the dummy variables / one-hot-encoding approach.  Also, for independent variables that display as vertical columns structure, instead of cloud structure, we will need to restructure the data into bins to help show a linear pattern between the predictors and target.


```python
#first, create bins for homes that have basements or do not have basements
bins = [ -1 , 1 , df.sqft_basement.astype('int').max() ] # to resolve error 'bins must increase monotonically.'
labels = ['0','1']
df.sqft_basement = pd.cut(df.sqft_basement, bins=bins,labels=labels)

#change to category data type
df.sqft_basement = df.sqft_basement.astype('category')

#Change the sqft_basement column over to dummies, remove the original column from the data, and add the dummy column
dumsqft_basement = pd.get_dummies(df.sqft_basement, prefix='basement', drop_first=True)
df = pd.concat([df, dumsqft_basement], axis = 1)
df.drop(['sqft_basement'], axis=1, inplace=True)
```

```python
#first, create bins for homes that are renovated or not renovated
bins = [ -1 , 1 , df.yr_renovated.astype('int').max() ]
labels = ['0','1']

df.yr_renovated = pd.cut(df.yr_renovated, bins=bins, labels=labels)

#change to category data type
df.yr_renovated = df.yr_renovated.astype('category')

#Change the yr_renovated column over to dummies, remove the original column from the data, and add the dummy column
dumyr_renovated = pd.get_dummies(df.yr_renovated, prefix='renovated', drop_first=True)
df = pd.concat([df,dumyr_renovated], axis = 1)
df.drop(['yr_renovated'], axis=1, inplace=True)
```

```python
#first, create bins for lat & long
bins = list(np.linspace(df.long.astype('int').min(), df.long.astype('int').max(), num=10))
labels = [i for i in range(1,10)]
df.lat = pd.cut(df.lat, bins, labels=labels)
df.long = pd.cut(df.long, bins, labels=labels)

#Change to category data type and change lat & long columns over to dummies 
df.lat = df.lat.astype('category')
dumlat = pd.get_dummies(df.lat,prefix='lat', drop_first=True)

df.long = df.long.astype('category')
dumlong = pd.get_dummies(df.long,prefix='long', drop_first=True)

# Adding dummy columns to the dataset, remove the original lat & long columns from the data
df = pd.concat([df,dumlat, dumlong], axis = 1)
df.drop(['lat','long'], axis=1, inplace=True)
```

### 4b. Feature Scaling & Normalization

In Section 3d <a href="#3d.-Normality-check">Normality check</a> performed a visual check and can tell that the shape of our variable's data distribution have different levels of skewness issues and does not follow a normal distribution. The assumption for multivariate normality is essential if we want to do statistical hypothesis testing to allow us to generate reliable confidence intervals and prediction intervals. We need to transform our data to make it Gaussian-like and perform feature scaling to manage the difference in magnitude.  
Feature scaling is a method used to standardize the range of independent variables/feature.  It is used to
scale the data in a specific range like -1 to +1 or -5 to +5 depending on the algorithm.If the data is not scaled or not normalized then we risk not getting the desired result and the result will be misleading. In addition if the data lies between the different mean values, it will result in two different standard deviations and variance so in order to achieve the expected output we need to transform the data in a similar range.


```python
# Perform logarithmic transformation to reduce skewness and decrease variability of data
df_log = pd.DataFrame([])
df_log['price'] = np.log(df['price'])
df_log['sqft_living']  = np.log(df['sqft_living'])
df_log['sqft_living15']  = np.log(df['sqft_living15'])
df_log['sqft_lot']  = np.log(df['sqft_lot'])

#Let's inspect the results of the log transformed features

df_log.hist(figsize  = [12, 8]);

plt.figure(figsize=(12,5))
sns.distplot(df_log.price, label='price')
sns.distplot(df_log.sqft_living, label='Sqft Living')
sns.distplot(df_log.sqft_living, label='Sqft Living15')
sns.distplot(df_log.sqft_lot, label='Sqft Lot')

plt.title('Log Transformed Features', fontdict={'fontsize': 20})
plt.xlabel('Features')
plt.legend()
plt.show()
```

```python
#We can see an improvement on the skewness
logprice = df_log['price'] 
logsqft_living = df_log['sqft_living']  
logsqft_living15 = df_log['sqft_living15']  
logsqft_lot = df_log['sqft_lot']  

# Let's perform Min Max Scaling to bring the values between 0 and 1 to manage the difference in magnitude

df['price'] = ( logprice - min(logprice) ) / ( max(logprice) - min(logprice) )
df['sqft_living'] = ( logsqft_living - min(logsqft_living) ) / ( max(logsqft_living) - min(logsqft_living) )
df['sqft_living15'] = ( logsqft_living15 - min(logsqft_living15) ) / ( max(logsqft_living15) - min(logsqft_living15) )
df['sqft_lot'] = ( logsqft_lot - min(logsqft_lot) ) / ( max(logsqft_lot) - min(logsqft_lot) )

data_cont_scaled = pd.DataFrame([])
data_cont_scaled['price']= df['price'] 
data_cont_scaled['sqft_living'] = df['sqft_living'] 
data_cont_scaled['sqft_living15'] = df['sqft_living15']
data_cont_scaled['sqft_lot']  = df['sqft_lot']

data_cont_scaled.hist(figsize  = [16, 6]);

plt.figure(figsize=(12,5))
sns.distplot(data_cont_scaled.price, label='price')
sns.distplot(data_cont_scaled.sqft_living, label='Sqft Living')
sns.distplot(data_cont_scaled.sqft_living, label='Sqft Living15')
sns.distplot(data_cont_scaled.sqft_lot, label='Sqft Lot')

plt.title('Scaled Features', fontdict={'fontsize': 20})
plt.xlabel('Features')
plt.legend()
plt.show()
```

MSE’s value is in helping us to select one prediction model over another.

In this regression project, we try to measure the model parameters (coefficients ) so our null and alternative hypotheses can be accepted or rejected.

NULL HYPOTHESIS (Ho): There is no relationship between Home Sale price and location(zip), sqft living, waterfront, views, condition, etc..


ALTERNATIVE HYPOTHESIS (Ha): There is "some" relation between Home Sale Price and and location(zip), sqft living, waterfront, views, condition, etc.

Rejecting the null hypothesis would help us associate some significance with our results
As per out model results, we reject the null (and thus believe the alternative) if the 95% confidence interval does not include zero for co-efficient. The p-value represents the probability that the coefficient is actually zero.  

A low R-squared value means that the predictors in oour model do not describe the variation in our dependent variable.  This project demonstrated that there are strong relationships between locations (zip/waterfront) and sq ft living area.

R-squared or coefficient of determination is a statistical measure that is used to assess the goodness of fit of a regression model.

R-squared uses a baseline model which is the **worst** model because it does not make use of any independent variables to predict the value of dependent variable Y, but rather it uses the **mean** of the observed responses of dependent variable Y and always predicts this mean as the value of Y. We looked at the adjusted R squared to keep our model in check as this modified R squared will increase only if the new terms improves our model.
